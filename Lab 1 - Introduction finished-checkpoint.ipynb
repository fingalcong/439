Suppose we want to visualize the function $g(t) = a \cdot \sin(2 \pi f t)$ while varying the values $f, a$. Generate a 2 by 2 plot that plots the function $g(t)$ as a line plot with values $f = 2, 8$ and $a = 2, 8$. Set the x limit of all figures to $[0, \pi]$ and the y limit to $[-10, 10]$. The figure size should be 8 by 8. Make sure to label your x and y axes with the appropriate value of $f$ or $a$. Additionally, make sure the x ticks are labeled $[0, \frac{\pi}{2}, \pi]$. Your overall plot should look something like this:

## Question 2.3 3D Plotting
The code below plot function f(x,y) in 3D. The function is 
$$f(x,y) = x^2 + y^2$$
### Activity 1
Observe the values of (x,y) that minimizes the function f(x,y)

--- Your answer here ----
### Activity 2
Change the function to $$f(x,y) = xy $$ and re-plot. Estimate the values (x,y) that minimizes f(x,y)?

--- Your answer here ----

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x**2 + y**2)

def fr(x,y):
    return (x*y)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

Z1 = fr(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z1, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

We should also familiarize ourselves with looking up documentation and learning how to read it. Below is a section of code that plots a basic wireframe. Replace each `# Your answer here` with a description of what the line above does, what the arguments being passed in are, and how the arguments are used in the function. For example your answer may look like,

Hint: The `Shift` + `Tab` (to see what a function does) tip from earlier in the notebook may help here.

```python
from mpl_toolkits.mplot3d import axes3d

u = np.linspace(1.5*np.pi, -1.5*np.pi, 100)
# Your answer here
# The array strat from 1.5pi and stop at -1.5pi, there are 100 of samples to generate.
[x,y] = np.meshgrid(u, u)
# Your answer here
# The meshgrid function expands two input arrays u and u. u is expanded with respect to the latter, and u is expanded with respect to u, u vertically and u horizontally.
squared = np.sqrt(x.flatten()**2 + y.flatten()**2)
z = np.cos(squared)
# Your answer here
# Cosine np.sqrt(x.flatten()**2 + y.flatten()**2).
z = z.reshape(x.shape)
# Your answer here
# Gives a new shape to an array without changing its data.

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Your answer here
# This adds an subplot figure with 1 row, 1 column,1 index and 3d projection.
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
# Your answer here
# Plot a 3D wireframe with 10 row stride and 10 column stride.
ax.view_init(elev=50., azim=30)
# Your answer here
# Set the elevation of axes in 50 degrees and azimuth of the axes in 30 degrees. This can be used to rotate the axes programatically.
plt.savefig("figure1.png")
```
